Linear model Classification
분류 모델 : SVM, Logistic Regression
Linear model은 회귀 뿐만 아니라 분류에도 사용되는데
Linear model의 회귀는  Linear, Lasso, Ridge가 있다.
회귀와 분류의 차이점은 수식상의 차이점은 아래와 같다.
y = w[0]*x[0] + w[1]*x[1]+ ... +w[p]*x[p] + b > 0 : 분류
y = w[0]*x[0] + w[1]*x[1]+ ... +w[p]*x[p] + b = 0 : 회귀
회귀는 추정되는 값을 찾는 것이고 분류는 class를 구분하는 것이다.


데이터셋 : make_forge()
사용 모델 : LinearSVC(suppoer vector Classifier), Logistic Regression

```python
import mglearn
from matplotlib import pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
X, y =mglearn.datasets.make_forge()
fig,axes = plt.subplots(1,2, figsize=(10,3))

for model, ax in zip([LogisticRegression(), LinearSVC()], axes):
    clf = model.fit(X,y)
    # fill : 배경색 채우기, eps : 여백, alpha : 분할 경계선의 투명도
    mglearn.plots.plot_2d_separator(clf,X,fill=False,eps=0.1,ax=ax,alpha=.8)
    mglearn.discrete_scatter(X[:,0],X[:,1],y,ax=ax)
    ax.set_title('{}'.format(clf.__class__.__name__))
    ax.set_xlabel('Feature0')
    ax.set_ylabel('Feature1')
    ax.legend()
plt.show()
```
