```python
#support
db = [['빵', '우유'], ['빵', '기저귀', '맥주', '달걀'], ['우유', '기저귀', '맥주', '콜라'],
 [ '우유','맥주', '기저귀','빵'], ['빵', '우유', '기저귀', '콜라']]
print('db =',db)

x = ['우유', '기저귀', '맥주']

cnt = 0 

for i in range(5):
    print('현 db[i]의 값', db[i] )
    print('x는', x)
    flag = False
    for j in x:
        print(x,'중', j)
        if j not in db[i]: #not in을 하는 이유 : 셋다 있어야 flag 가 False가 되고 그래야 if문이 돌아갈테니까
            flag = True
            print(j,'가 들어가서 flag가',flag)
        else:
            print(j,'가 들어가서', 'flag가',flag)
    
    print('결국 flag는',flag)
    if flag == False:
        cnt = cnt+1
        print("--"*30)
        print(db[i],'일때 총cnt값', cnt)
        print("--"*30)

s =cnt/len(db)
print()
print('총 cnt값은', cnt)
print('support(%s) = cnt/전체db수 = %s' %(x,s))
```

```python
'''confidence 우유, 기저귀 중 우유 기저귀 맥주가 포함된 비율'''

# {우유, 기저귀, 맥주} /{우유, 기저귀} 
db = [['빵', '우유'], ['빵', '기저귀', '맥주', '달걀'], ['우유', '기저귀', '맥주', '콜라'],
 [ '우유','맥주', '기저귀','빵'], ['빵', '우유', '기저귀', '콜라']]

x = ['우유', '기저귀', '맥주']

#{우유, 기저귀, 맥주} 사실 이전 support 구할 때랑 동일하다.
c2 = 0
for i in range(5):
    if checkSupport(x, db) == 1:
        c2 += 1
        print('우유, 기저귀, 맥주가 포함된 레코드', db[i])
#{우유, 기저귀}
c1 = 0 
x = x[0:2]
for i in range(5):
    if checkSupport(x,db) == 1:
        c1 += 1
        print('우유, 기저귀가 포함된 레코드', db[i])



confidence = c2/c1  # {우유, 기저귀, 맥주} /{우유, 기저귀}
print(' {우유, 기저귀, 맥주} /{우유, 기저귀} =', c1,'/',c2) 
print('구하고자 하는 confidence 값은 %.2f' %confidence)
```

```python
#유클리드 거리 (점과 점 사이의 거리)

dic = {'김영희' : [23,2,3] , '배철수' : [40,10,1]}
print(dic)
#딕셔너리에 각 점들을 대입해준다.
#print(dic['김영희'][1]) # 2

'''구상 : 3번 돌면서 각 인덱스에 접근 후 빼고 제곱해준다. 그것들을 다 더해주고(sum)
마지막에 루트를 씌어준다.'''
sum = 0
for i in range(3):
    #접근
    s = (dic['김영희'][i] - dic['배철수'][i])**2
    sum  = sum + s

ans = sum**0.5

print('구하고자 하는 유클리드 거리는 %.3f' %ans)
```

```python
#jaccard similarity
#구하고자 하는 값들을 딕셔너리의 value값에 넣는다.
dic = {
    'ullman':[('jeff', 'ullman', 'jennifer', 'widom', 'xml', 'security', 'vldb04'), 
 ('jeff', 'ullman', 'hector', 'garcia-molina', 'hashing', 'vldb04'), \
 ('jeff', 'ullman', 'xml', 'query', 'vldb04'), \
 ('jeff', 'ullman', 'xml', 'vldb06')],
 'widom':[('hector', 'garcia-molina', 'jennifer', 'widom', 'jeff', 'ullman', \
 'ranking', 'vldb04')]
 }

'''구상 :
자카드 유사도와 같은 경우는 자료들을 전부 set에 집어넣어야 한다. (자카드 유사도가 
집합으로 구성되어있기 때문.)
set1 = ullman
set2 = widom
set1 과 set2의 교집합을 구한다. (if x in y : count ++;)
len(set1) - len(set2) + 교집합 수 = 전체 집합 수
자카드 유사도를 구한다. 
교집합 수 / 전체 집합 수
'''

'''2차원 배열이므로 이중 for문을 사용하여 전부 접근한다.'''

ullman = set()
for i in range(len(dic['ullman'])):
    for k in range(len(dic['ullman'][i])):
        j = dic['ullman'][i][k]
        ullman.add(j)

widom = set()
for i in range(len(dic['widom'])):
    for k in range(len(dic['widom'][i])):
        j = dic['widom'][i][k]
        widom.add(j)

count = 0
for i in ullman:     # i <- ullman
    if i in widom:
        count = count + 1

total = len(ullman) + len(widom) - count        

print(ullman)
print(widom)
print('둘의 교집합 수는', count)
print('따라서 구하고자 하는 자카드 유사도는 %.3f' %(count / total))
print('자카드 거리는 %.3f' %(1-(count / total))) 
```

```python
#편집거리

a = 'strong'
b = 'stone'


#빈 배열을 생성해준다. len_a * len_b
#이차원 배열이니 당연히 이중 for문으로

'''
  0 s t o n e
  s
  t
  r
  o
  n
  g

'''
list = [[0]*(len(b)+1)
        for i in range(len(a)+1)]
#print(list)

'''   s t o n e
    0 1 2 3 4 5 
  s 1
  t 2
  r 3
  o 4
  n 5
  g 6
  '''

#초기화 해준다.
for i in range(len(b) + 1):
    list[0][i] = i
    
for i in range(len(a) + 1):
    list[i][0] = i
    
print(list)

for i in range(1, len(a)+1):
    for j in range(1,len(b)+1):

        if(a[i-1] == b[j-1]):
            list[i][j] = min(list[i-1][j]+1, list[i][j-1]+1, list[i-1][j-1]+0)
        else:
            list[i][j] = min(list[i-1][j]+1, list[i][j-1]+1, list[i-1][j-1]+1)

print(list)
print('edit distance (%s, %s) = %d' %(a,b,list[len(a)][len(b)]))
```

```python
def edit_distance(a,b):
    list = [[0]*(len(b)+1)
        for i in range(len(a)+1)]


    #초기화 해준다.
    for j in range(len(b) + 1):
        list[0][j] = j

    for i in range(len(a) + 1):
        list[i][0] = i

   
    for i in range(1, len(a)+1):
        for j in range(1,len(b)+1):

            if(a[i-1] == b[j-1]):
                list[i][j] = min(list[i-1][j]+1, list[i][j-1]+1, list[i-1][j-1]+0)
            else:
                list[i][j] = min(list[i-1][j]+1, list[i][j-1]+1, list[i-1][j-1]+1)

    return list, list[len(a)][len(b)]
    
```

```python
# 코사인 유사도  #겹치는 단어 개수에 따른 벡터 사이각 계산
import pandas as pd

# Tom_Cruise와 T_Cruise의 유사도
dic = {'Tom_Cruise': [('collateral', '04'), ('last', 'samurai', '03'),
                      ('minority', 'report', '02'), ('vanilla', 'sky', '02')],
       'T_Cruise': [('vanilla', 'sky'), ('the', 'last', 'samurai'),
                    ('mission', 'impossible'), ('minority', 'report')]
       }

'''리스트를 만든다. -> 단어 수에 해당하는 딕셔너리를 만든다.
널값을 0으로 하고 두 딕셔너리를 합한다. (판다스 이용)
코사인 유사도를 구한다.
'''

print("단어들을 리스트로 구성")

g1 = []
for i in range(len(dic['Tom_Cruise'])):
    for j in range(len(dic['Tom_Cruise'][i])):
        g1.append(dic['Tom_Cruise'][i][j])
print(g1)

g2 = []
for i in range(len(dic['T_Cruise'])):
    for j in range(len(dic['T_Cruise'][i])):
        g2.append(dic['T_Cruise'][i][j])
print(g2)

print("딕셔너리로 각 단어 수 계산")

Tom_dic = dict()
T_dic = dict()

for word in g1:
    if word not in Tom_dic:
        Tom_dic[word] = 1
    else:
        Tom_dic[word] = Tom_dic[word] + 1

for word in g2:
    if word not in T_dic:
        T_dic[word] = 1
    else:
        T_dic[word] += 1

print(Tom_dic)
print(T_dic)

'''합 집합을 만든다.'''



summary = pd.DataFrame({
    'Tom_Cruise': Tom_dic,
    'T_Cruise': T_dic
})
summary['Tom_Cruise'] = summary['Tom_Cruise'].fillna(0)
summary['T_Cruise'] = summary['T_Cruise'].fillna(0)

print(summary)


#분모
sum_Tom = 0
for i in range(len(summary)):
    sum_Tom += summary['Tom_Cruise'][i] ** 2

sum_Tom = sum_Tom ** 0.5
print('TomCruise의 벡터길이 =', sum_Tom)

sum_T = 0
for i in range(len(summary)):
    sum_T += summary['T_Cruise'][i] ** 2

sum_T = sum_T ** 0.5
print('TCruise의 벡터길이 =', sum_T)

#분자

up = 0
for i in range(len(summary)):
    up += summary['Tom_Cruise'][i] * summary['T_Cruise'][i]


Cosine_similarity =  up  / (sum_Tom * sum_T)
Cosine_distance = 1 - Cosine_similarity

print('Cosine_similarity = %.4f' %Cosine_similarity)
print('Cosine distance =%.4f' %Cosine_distance)


print('Tom_Cruise의 02항을 1로 바꿨을 때 코사인 유사도를 구해보자')
summary['Tom_Cruise']['02'] = 1.0

sum_Tom = 0
for i in range(len(summary)):
    sum_Tom += summary['Tom_Cruise'][i] ** 2

sum_Tom = sum_Tom ** 0.5
print('TomCruise의 벡터길이 =', sum_Tom)

sum_T = 0
for i in range(len(summary)):
    sum_T += summary['T_Cruise'][i] ** 2

sum_T = sum_T ** 0.5
print('TCruise의 벡터길이 =', sum_T)

#분자

up = 0
for i in range(len(summary)):
    up += summary['Tom_Cruise'][i] * summary['T_Cruise'][i]


Cosine_similarity =  up  / (sum_Tom * sum_T)
Cosine_distance = 1 - Cosine_similarity

print('Cosine_similarity = %.4f' %Cosine_similarity)
print('Cosine distance =%.4f' %Cosine_distance)
```
